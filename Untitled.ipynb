```python
import torch
import numpy as np
import cv2
from ultralytics import YOLO
from matplotlib import pyplot as plt
```

```python
# Run detection on an online image
# results = model.predict(source="https://ultralytics.com/images/zidane.jpg", show=True)
```

```python
# load model
model = YOLO("yolov8n.pt")
```

```python
model
```

```python
img = "https://ultralytics.com/images/zidane.jpg"
cars = "https://cdn.standardmedia.co.ke/images/tuesday/evingu8un5xyhs855d493f6df4084.jpg"
```

```python
results = model(cars)[0]
results.verbose()
```

```python
annotated_img = results.plot()
plt.imshow(annotated_img)
plt.axis("off")
plt.show()
```

```python
plt.imshow(np.squeeze(results.plot()))
```

```python
# real time detection
cap = cv2.VideoCapture(0, cv2.CAP_AVFOUNDATION)
while cap.isOpened():
    ret, frame = cap.read()

    # make detection
    results = model(frame)

    annotated_frame = results[0].plot()
    cv2.imshow("YOLO", annotated_frame)

    if cv2.waitKey(10) & 0xFF == ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```
